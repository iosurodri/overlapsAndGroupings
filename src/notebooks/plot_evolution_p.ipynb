```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
PATH_logs = os.path.join('..', '..', 'reports', 'logs')
data_names = []
data = []

for log_file in os.listdir(PATH_logs):
    data_names.append(log_file)
    data.append(pd.read_csv(os.path.join(PATH_logs, log_file)))
```

```python
logs = []
for log in data:
    logs.append(log['Value'].to_numpy())
```

```python
fig = plt.figure()
for log in logs:
    plt.plot(log)
plt.xlabel('Epoch')
plt.ylabel(r'$p$')
plt.title(r'Evolution of parameter $p$')
```

```python
PATH_plots = os.path.join('..', '..', 'reports', 'plots')
# plt.savefig(os.path.join(PATH_plots, 'p_evolution.pdf'))

fig.savefig(os.path.join(PATH_plots, 'p_all_evolution.pdf'), bbox_inches="tight")
```
